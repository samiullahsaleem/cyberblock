```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, marker = '+', color = 'red')
```

```python
area = df.drop('price', axis = 1)
area.shape
```

```python
reg  = linear_model.LinearRegression()
reg.fit(area, df.price)
```

```python
reg.predict([[3300]])
```

```python
c = reg.coef_
c
```

```python
intercept = reg.intercept_
intercept
```

```python
c*3300+intercept
```

```python
area2 = pd.read_csv('areas.csv')
area2
```

```python
p = reg.predict(area2)
p
```

```python
area2['prediction_price'] = p
```

```python
area2.to_csv('prediction_price.csv', index = False)
```
