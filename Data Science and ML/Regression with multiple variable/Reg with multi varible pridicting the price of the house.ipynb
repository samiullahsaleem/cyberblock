```python
import pandas as pd
import numpy as np
from sklearn import linear_model 
```

```python
df = pd.read_csv('homeprices.csv')
df.head()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df.head(5)
```

```python
x = df.drop('price', axis= 1)
x
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, df.price)
```

```python
reg.predict([[3000,4,3]])
```

```python
# here we have three independent variables so we got three coeffiecient for each
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
3000*m[0]+4*m[1]+3*m[2]+b
```
