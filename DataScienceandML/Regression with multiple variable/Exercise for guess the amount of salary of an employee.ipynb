```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib as plt
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience = df.experience.fillna("Zero")
df1= df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
df1
```

```python
import math
scores = math.floor(df['test_score(out of 10)'].mean())
scores
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(scores)
df
```

```python
from word2number import w2n 
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
x = df.drop('salary($)', axis = 1)
x
```

```python
y = df['salary($)']
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
reg.predict([[2,9,6]])
```
